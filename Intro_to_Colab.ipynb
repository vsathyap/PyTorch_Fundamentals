# **Tutorial 3: Tensor Operations**

```python
import torch
```

```python
# to create a tensor from 0->n-1 use arange(n)
t1  = torch.arange(10)
t1
```

## 'RESHAPE' & 'VIEW'

```python
# Using Reshape
t1 = t1.reshape(2,5)
t1
```

```python
# Reshape when we don't know the number of items
t2 = torch.arange(10)
print(f"t1 = {t1}")
t2 = t2.reshape(2,-1)
print(f"t2 = {t2}")
t2 = t2.reshape(-1,10)
print(f"t2 = {t2}")
```

```python
t3 = torch.arange(10)
t3
```

```python
# Using View
t3 = t3.view(2,5)
t3
```

```python
# With reshape and view, the variables will update
t5 = torch.arange(10)
print(f"Original t5 = {t5}")
t6 = t5.reshape(5,-1)
print(f"Original t6 = {t6}")
t5[0] = 999
print(f"Modified t5 = {t5}")
print(f"Modified t6 = {t6}")
```

```python
# Grab a slice of a tensor
t7 = torch.arange(8)
print(f"t7 = {t7}")
t7 = t7.reshape(4,-1)
print(f"Reshaped t7 = {t7}")
print(f"t7[:,1] = {t7[:,1]}")
print(f"To retain the column shape use, t7[:,1:] = {t7[:,1:]}")
```

# **Tutorial 4: Tensor Math Operations**

```python
import torch
import numpy as np
```

```python
t_a = torch.tensor([1,2,3,4])
t_b = torch.tensor([5,6,7,8])
```

```python
# Addition short-hand
t_a + t_b
```

```python
#Addition long-hand (using functions)
torch.add(t_a,t_b)
```

```python
# Subtraction
t_a - t_b
```

```python
#Subtraction long-hand (using functions)
torch.sub(t_a,t_b)
```

```python
#Multiplication
t_a*t_b
```

```python
#Multiplication long-hand (using functions)
torch.mul(t_a,t_b)
```

```python
# Division
t_a/t_b
```

```python
#Division long-hand (using functions)
torch.div(t_a,t_b)
```

```python
#Remainder
t_a % t_b
```

```python
#Remainder long-hand (using functions)
torch.remainder(t_a,t_b)
```

```python
#Exponent
t_a**t_b
```

```python
#Exponential long-hand (using functions)
torch.pow(t_a,t_b)
```

```python
# Reassignment
t_a = t_a + t_b
t_a
```

```python
t_a.add_(t_b)
```
